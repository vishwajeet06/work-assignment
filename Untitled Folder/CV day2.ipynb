```python
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('external-content.duckduckgo.com.jpg')
```

```python
print(image.shape)
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.title("gray-img")
plt.imshow(gray_img)

plt.subplot(1,2,2)
plt.title("gray-img")
plt.imshow(image)

plt.show()
```
